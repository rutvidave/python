```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv("Titanic-Dataset.csv")
df
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
df.drop(columns=['Cabin'], inplace=True)
```

```python
print("Missing Values After Cleaning:")
print(df.isnull().sum())
```

```python
df.describe(include='all')
```

```python
df.head()
```

```python
df.tail()
```

```python
plt.figure(figsize=(5, 4))
df['Survived'].value_counts().plot(kind='bar', color=['skyblue', 'pink'])
plt.title('Survival Count')
plt.xlabel('Survived')
plt.ylabel('Count')
plt.xticks([0, 1], ['No', 'Yes'])
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
df['Age'].plot(kind='hist', bins=30, color='lightblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
df['Fare'].plot(kind='hist', bins=30, color='lightpink', edgecolor='black')
plt.title('Fare Distribution')
plt.xlabel('Fare')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
pd.crosstab(df['Sex'], df['Survived']).plot(kind='bar', stacked=True, color=['lightblue', 'lightpink'])
plt.title('Survival by Sex')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
pd.crosstab(df['Pclass'], df['Survived']).plot(kind='bar', stacked=True, color=['lightblue', 'lightpink'])
plt.title('Survival by Passenger Class')
plt.xlabel('Pclass')
plt.ylabel('Count')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

 The chart shows that 1st class passengers survived more often than those in lower classes.

```python
plt.figure(figsize=(6, 4))
plt.scatter(df['Age'], df['Fare'], c=df['Survived'], cmap='coolwarm', alpha=0.6)
plt.title('Age vs Fare (Color = Survival)')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
corr = df.corr(numeric_only=True)
fig, ax = plt.subplots(figsize=(6, 5))
cax = ax.matshow(corr, cmap='coolwarm')
fig.colorbar(cax)
plt.xticks(range(len(corr.columns)), corr.columns, rotation=90)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title("Correlation Heatmap", pad=20)
plt.tight_layout()
plt.show()
```

```python
title_counts = df['Title'].value_counts()
survived_by_title = df.groupby(['Title', 'Survived']).size().unstack(fill_value=0).loc[title_counts.index]
survived_by_title.plot(kind='bar', stacked=True, figsize=(8, 4),color = ["lightblue","lightpink"])
plt.title('Survival Count by Title')
plt.xlabel('Title')
plt.ylabel('Count')
plt.xticks(rotation=60)
plt.show()
```

```python
df["FamilySize"] = df["SibSp"] + df["Parch"] + 1

df.groupby("FamilySize")["Survived"].value_counts().unstack().plot(kind="bar", figsize=(8,4),color = ["lightblue","lightpink"])
plt.title("Survival by Family Size")
plt.xlabel("Family Size")
plt.ylabel("Count")
plt.show()
```

```python
df.groupby("Survived")["Age"].mean().plot(kind="bar",color = "lightblue")
plt.title("Average Age: Survived vs Not Survived")
plt.xlabel("Survived")
plt.ylabel("Average Age")
plt.show()
```

```python
df["AgeGroup"] = pd.cut(df["Age"], bins=[0, 12, 18, 30, 50, 80], 
                        labels=["Child", "Teen", "Young Adult", "Adult", "Senior"])
fare_by_age = df.groupby("AgeGroup")["Fare"].mean()

fare_by_age.plot(kind="line", marker="o")
plt.title("Average Fare by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Fare")
plt.show()
```

```python
df["Sex"].value_counts().plot(kind="pie", autopct="%1.1f%%", startangle=90,colors = ["lightblue", "lightpink"])
plt.title("Gender Distribution")
plt.ylabel("")
plt.show()
```

```python
age_bins = [0, 12, 18, 30, 45, 60, 100]
age_labels = ['Child', 'Teen', 'Young Adult', 'Adult', 'Middle Age', 'Senior']
df['AgeGroup'] = pd.cut(df['Age'], bins=age_bins, labels=age_labels)
```

```python
age_survival = df.groupby('AgeGroup')['Survived'].mean()
age_survival.plot(kind='bar', color='skyblue')
plt.title("Survival Rate by Age Group")
plt.ylabel("Survival Rate")
plt.xlabel("Age Group")
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
df['TicketFreq'] = df.groupby('Ticket')['Ticket'].transform('count')
```

```python
df['GroupTicket'] = df['TicketFreq'] > 1
group_survival = df.groupby('GroupTicket')['Survived'].mean()
```

```python
group_survival.plot(kind='bar', color=['skyblue', 'pink'])
plt.xticks([0,1],['Solo Traveler', 'Group Ticket'])
plt.ylabel('Survival Rate')
plt.title('Survival Rate: Solo vs Group Ticket Holders')
plt.grid(axis='y')
plt.show()
```

This dataset contains information about 891 passengers who were on the Titanic. It includes their age, gender, ticket price, class, family members onboard, and whether or not they survived.



- Only 38% of the people survived.That means almost 6 out of 10 passengers did not make it.

- Women had a much higher chance of survival.Most men did not survive

- There were 3 classes: 1st, 2nd, and 3rd.
    1st class passengers were most likely to survive.
    People in 3rd class had the lowest survival rate.

- The average age was around 29 years.The youngest passenger was less than 1 year old.
The oldest was 80.
Children had slightly better chances of survival, especially if traveling with family.

- Ticket prices ranged from free (or very cheap) to over $500.
People who paid more (mostly 1st class) had better survival chances.

- Most passengers got on the ship from Southampton (UK).
Others boarded in Cherbourg (France) and Queenstown (Ireland).

- Most passengers were traveling alone.
Some had large families.
Having family slightly improved survival chances, especially for women and children.

- Some passengers were missing age or boarding port info.
These were filled in using the most common value.
One column (Cabin) was removed because it had too much missing data.
